To get the link to a file stored in GitHub

- click on View Raw,
- copy the URL of the raw file, 
- use this URL as the location of your file.

```python
import pandas as pd
shelters2017 = pd.read_csv('https://raw.githubusercontent.com/oke-w/shelters/main/daily-shelter-occupancy-2017.csv')
shelters2018 = pd.read_csv('https://raw.githubusercontent.com/oke-w/shelters/main/daily-shelter-occupancy-2018.csv')
shelters2019 = pd.read_csv('https://raw.githubusercontent.com/oke-w/shelters/main/daily-shelter-occupancy-2019.csv')
shelters2020 = pd.read_csv('https://raw.githubusercontent.com/oke-w/shelters/main/daily-shelter-occupancy-202010.csv')
```

```python
shelters2017.head(10)
```

```python
shelters2020.info()
shelters2020.drop(columns=['id','DateRange'], inplace=True)
```

```python
shelters = pd.concat([shelters2017, shelters2018, shelters2019, shelters2020])
shelters.info()
```

The info() shows that there are a very small fraction of the Postal Codes and Capacity values that are null for 2017, 2018, and 2019 data.

The 'Occupancy Date' column is storing Strings ... it should probably be converted to a date format.

```python
shelters2017_by_date = shelters2017.groupby('OCCUPANCY_DATE')[['OCCUPANCY','CAPACITY']].sum()
shelters2018_by_date = shelters2018.groupby('OCCUPANCY_DATE')[['OCCUPANCY','CAPACITY']].sum()
shelters2019_by_date = shelters2019.groupby('OCCUPANCY_DATE')[['OCCUPANCY','CAPACITY']].sum()
shelters2020_by_date = shelters2020.groupby('OCCUPANCY_DATE')[['OCCUPANCY','CAPACITY']].sum()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = [12,4])
ax = fig.add_subplot(1, 1, 1)
plt.title("Shelter Occupancy by Year")
ax.plot(range(365),shelters2017_by_date.OCCUPANCY, '-', c='red')
ax.plot(range(365),shelters2018_by_date.OCCUPANCY, '-', c='green')
ax.plot(range(365),shelters2019_by_date.OCCUPANCY, '-', c='blue')
ax.plot(range(len(shelters2020_by_date.OCCUPANCY)),shelters2020_by_date.OCCUPANCY, '-', c='orange')
```

Not sure what's going on with 2017 (red) data.  Would have to investigate.
The 2018 data (green) seems to stabilize halfway through the year - maybe more homes were being added to the stats.
The 2019 data (blue) drops by a few hundred at the end of Aug - maybe reporting or home closures?
The 2020 data (orange) shows a significant drop at the end of Mar, and another blip in early Oct.

Nothing seems to show a seasonal fluctuation, although by cleaning and filtering the data, this might become evident.

```python
shelters2017_by_date['AVAILABILITY'] = shelters2017_by_date.CAPACITY - shelters2017_by_date.OCCUPANCY
shelters2017_by_date['AVAILABILITY_PCT'] = 100 * shelters2017_by_date.AVAILABILITY / shelters2017_by_date.CAPACITY
shelters2018_by_date['AVAILABILITY'] = shelters2018_by_date.CAPACITY - shelters2018_by_date.OCCUPANCY
shelters2018_by_date['AVAILABILITY_PCT'] = 100 * shelters2018_by_date.AVAILABILITY / shelters2018_by_date.CAPACITY
shelters2019_by_date['AVAILABILITY'] = shelters2019_by_date.CAPACITY - shelters2019_by_date.OCCUPANCY
shelters2019_by_date['AVAILABILITY_PCT'] = 100 * shelters2019_by_date.AVAILABILITY / shelters2019_by_date.CAPACITY
shelters2020_by_date['AVAILABILITY'] = shelters2020_by_date.CAPACITY - shelters2020_by_date.OCCUPANCY
shelters2020_by_date['AVAILABILITY_PCT'] = 100 * shelters2020_by_date.AVAILABILITY / shelters2020_by_date.CAPACITY
```

```python
fig = plt.figure(figsize = [12,4])
ax = fig.add_subplot(1, 1, 1)
plt.title("Shelter Availability (%) by Year")
ax.plot(range(365),shelters2017_by_date.AVAILABILITY_PCT, '-', c='red')
ax.plot(range(365),shelters2018_by_date.AVAILABILITY_PCT, '-', c='green')
ax.plot(range(365),shelters2019_by_date.AVAILABILITY_PCT, '-', c='blue')
ax.plot(range(len(shelters2020_by_date.AVAILABILITY_PCT)),shelters2020_by_date.AVAILABILITY_PCT, '-', c='orange');
```
